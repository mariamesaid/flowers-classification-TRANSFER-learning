```python
import tensorflow as tf
import tensorflow_hub as hub
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
import numpy as np

device_name = tf.test.gpu_device_name()
if device_name != '/device:GPU:0':
  print('GPU device not found. Check your settings!')
else:
  print('Found GPU at: {}'.format(device_name))
```

```python
(train_ds, val_ds), ds_info = tfds.load(
    'tf_flowers',
    split=['train[:80%]', 'train[80%:]'],
    as_supervised=True,
    with_info=True
)

num_classes = ds_info.features['label'].num_classes
class_names = ds_info.features['label'].names

def format_image(image, label):
    image = tf.image.resize(image, (224, 224)) / 255.0
    return image, label

BATCH_SIZE = 32
train_batches = train_ds.shuffle(1000).map(format_image).batch(BATCH_SIZE).prefetch(1)
val_batches = val_ds.map(format_image).batch(BATCH_SIZE).prefetch(1)
```

```python

base_model = tf.keras.applications.MobileNetV2(
    input_shape=(224, 224, 3),
    include_top=False,
    weights='imagenet'
)

base_model.trainable = False

model = tf.keras.Sequential([
    base_model,
    tf.keras.layers.GlobalAveragePooling2D(),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(num_classes, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
history = model.fit(train_batches, epochs=5, validation_data=val_batches)
```

```python
image_batch, label_batch = next(iter(val_batches))
predictions = model.predict(image_batch)
predicted_ids = np.argmax(predictions, axis=-1)

plt.figure(figsize=(10,9))
for n in range(9):
  plt.subplot(3,3,n+1)
  plt.subplots_adjust(hspace=0.3)
  plt.imshow(image_batch[n])
  color = "green" if predicted_ids[n] == label_batch[n] else "red"
  plt.title(f"Pred: {class_names[predicted_ids[n]]}\nActual: {class_names[label_batch[n]]}", color=color)
  plt.axis('off')
```
